### a from normal distribution

```python
process.data <- function(path, prefix) {
    power = data.frame()
    for (r in c(0.03, 0.05, 0.07, 0.10)) {
        files = dir(path, pattern = paste(prefix, paste("causal", r, sep = ""), sep = "."), full.names = T)
        raw.data = rbindlist(lapply(files, fread))
        raw.data$`ACAT-O` = apply(raw.data, 1, function(x) suppressWarnings(ACAT(c(x[4], x[5], x[6]))))
        dfrow = c(colSums(raw.data < 0.05) / nrow(raw.data), r)
        names(dfrow) = c("CMC", "WSS", "BURDEN", "VT", "SKAT-O", "ACAT-V", 
                           "LRT-q", "ACAT-O", "causal")
        power = bind_rows(power, dfrow)
    }
    power
}
```

```python
# a_from_normal_dist, 110var
a.norm.dist.110var = process.data("./data/simulation/power_other_settings/a_from_normal_dist", "p.values.acat")
```

```python
# a_from_normal_dist, 71var
a.norm.dist.71var = process.data("./data/simulation/power_other_settings/a_from_normal_dist", "p.values.acat.71var")
```

### Different variance

```python
process.data2 <- function(path, prefix) {
    power = data.frame()
    for (a in c(0.3, 0.6, 0.9, 1.2, 1.5)){
        for (r in c(0.03, 0.05, 0.07, "0.10")) {
            files = dir(path, pattern = paste(prefix, paste("causal", r, ".a", a, sep = ""), sep = "."), full.names = T)
            raw.data = na.omit(rbindlist(lapply(files, fread)))
            raw.data$`ACAT-O` = apply(raw.data, 1, function(x) suppressWarnings(ACAT(c(x[4], x[5], x[6]))))
            dfrow = c(colSums(raw.data < 0.05) / nrow(raw.data), r, a)
            names(dfrow) = c("CMC", "WSS", "BURDEN", "VT", "SKAT-O", "ACAT-V", 
                               "LRT-q", "ACAT-O", "causal", "a")
            power = bind_rows(power, dfrow)
        }
    }
    power
}
```

```python
# fixed a, 110 var
fixed.a.110var = process.data2("./data/simulation/power_other_settings/different_variance/", "p.values.fixed_a.110var")
```

```python
# fixed a, 71 var
fixed.a.71var = process.data2("./data/simulation/power_other_settings/different_variance/", "p.values.fixed_a.71var")
```

#### Calculate average number of rare variants

```python
cosi.hap = fread("./data/simulation/haplotype/len3k_71var/preprocessed.3k.hap.csv")
```

```python
repeats = 10000
individual = 1000
maf.cutoff = 0.05
num = rep(0, repeats)
for (i in 1:repeats) {
    hap4k <- cosi.hap[sample(nrow(cosi.hap), 2*individual),]
    hap2k <- hap4k[1:individual, ] + hap4k[(individual+1):(2*individual), ]
    hap2k <- as.matrix(hap2k)
    maf.list <- colMeans(hap2k, na.rm = T) / 2
    hap2k[, maf.list > 0.5] <- 2 - hap2k[, maf.list > 0.5]
    hap2k <- hap2k[, maf.list != 0 & maf.list != 1]
    maf.list <- colMeans(hap2k, na.rm = T) / 2

    # only use rare variants
    rare.var.num = sum(maf.list < maf.cutoff)
    num[i] = rare.var.num
}
```
